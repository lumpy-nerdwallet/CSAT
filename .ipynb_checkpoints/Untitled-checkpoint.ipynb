```python
def isASCIIString(string):
    try:
        string.decode('ascii')
        return True
    except:
        return False


def replaceString(string):
    return string.replace("\\n", " ").replace("&amp;", "&").replace('&#039;', '\'').replace("&quot;", "\"").replace("&lt;", "<").replace("&gt;", ">").strip()
#spellcheck
```
