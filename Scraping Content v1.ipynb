```python
output = []
for url in urlList[0:100]:
    try:
        print(url)
        r = requests.get(url)
        if r.status_code == requests.codes.ok:
            tree = html.fromstring(r.content)
            content = [string.encode("UTF-8") for string in tree.xpath('//div[@property="articleBody"]/p/text()|//div[@property="articleBody"]/h3/text()|//div[@property="articleBody"]/p/*/text()')]
            if len(content) > 0 and not content[0].startswith("by:"): ## dealing with names
                content = " ".join(content[1:])
                output.append([url, r.status_code, r.history, content])
            else:
                output.append([url, r.status_code, r.history, None])
        else: 
            output.append([url, r.status_code, r.history, None])
    except:
        print("Error: " + url)
        output.append([url, None, None, None])
```
